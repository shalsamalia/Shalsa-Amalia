# BAB I. Business Understanding #

**A. Determine Business Objective**

**1. Project Background**

Salah satu penunjang keberlangsungan roda perusahaan pembiayaan baik lembaga Bank maupun Non-Bank yaitu berasal dari penerimaan pendapatan bunga yang di dapat melalui penyaluran dana dalam bentuk kredit, baik kredit konsumtif maupun kredit produktif. Tetapi permasalahan yang sering terjadi ketika penyaluran dana dalam bentuk kredit adalah seringkali para nasabah melalaikan tanggung jawabnya untuk memenuhi kewajiban atau membayar kreditnya di karenakan adanya faktor kesengajaan ataupun kondisi di luar kemampuan nasabah. Permasalahan ini sangat di khawatirkan oleh setiap perusahaan pembiayaan dikarenakan akan mengganggu kondisi keuangan perusahaan dan bahkan dapat mengakibatkan berhentinya kegiatan usaha perusahaan, karena dari bunga yang dibayarkan nasabah, perusahaan mendapatkan laba guna membiayai kegiatan operasional perusahaannya. Dengan adanya permasalahan kredit macet yang berpengaruh terhadap penerimaan pendapatan perusahaan, maka dapat berimbas meningkatnya nilai NPL (Non Performing Loan) yang artinya berdampak negatif terhadap penilaian kinerja perusahaan dalam menangani permasalahan kredit macet. Karena dari nilai NPL inilah dapat menjadi suatu indikator penilaian kinerja perusahaan dalam menyalurkan dananya dalam bentuk kredit.

**2. Objective**

*   menilai kemungkinan keberhasilan pembayaran kredit bagi pemohon pinjaman
*   memprediksi apakah pemohon akan melakukan pembayaran tepat waktu atas kewajiban kredit mereka

**3. Business Success Criteria**

*  Pengurangan Tingkat Gagal Bayar: Salah satu tujuan utamanya adalah untuk meminimalkan tingkat gagal bayar, yang mengukur persentase pinjaman atau pembayaran kredit yang tidak dilunasi oleh peminjam.
*  Profitabilitas: Hal ini dapat diukur dengan mengevaluasi pendapatan bersih yang dihasilkan dari pembayaran kredit yang berhasil dibandingkan dengan biaya yang terkait dengan pembayaran kredit atau pinjaman.
*  Manajemen Risiko: Mengurangi kemungkinan non-pembayaran dengan mengidentifikasi peminjam berisiko tinggi dan menerapkan strategi manajemen risiko yang tepat. Dalam hal ini dapat melibatkan penetapan batas kredit, penyesuaian suku bunga, atau penerapan kriteria persetujuan yang lebih ketat.
*  Kepuasan Pelanggan: Kepuasan pelanggan merupakan aspek penting dari keberhasilan pembayaran kredit. Memastikan bahwa pelanggan dapat melakukan pembayaran dengan nyaman dan tanpa kesulitan dapat berkontribusi pada loyalitas pelanggan jangka panjang dan persepsi merek yang positif.

```python
# Reduce Memory Usage

def reduce_mem_usage(df):

    start_mem = df.memory_usage().sum() / 1024**2
    print('Memory usage of dataframe is {:.2f} MB'.format(start_mem))

    for col in df.columns:
        col_type = df[col].dtype

        if col_type != object:
            c_min = df[col].min()
            c_max = df[col].max()
            if str(col_type)[:3] == 'int':
                if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)
                elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)
                elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)
                elif c_min > np.iinfo(np.int64).min and c_max < np.iinfo(np.int64).max:
                    df[col] = df[col].astype(np.int64)
            else:
                if c_min > np.finfo(np.float16).min and c_max < np.finfo(np.float16).max:
                    df[col] = df[col].astype(np.float16)
                elif c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
                    df[col] = df[col].astype(np.float32)
                else:
                    df[col] = df[col].astype(np.float64)
        else:
            df[col] = df[col].astype('category')

    end_mem = df.memory_usage().sum() / 1024**2
    print('Memory usage after optimization is: {:.2f} MB'.format(end_mem))
    print('Decreased by {:.1f}%'.format(100 * (start_mem - end_mem) / start_mem))

    return df
```
